# Importante

Seleção do Ano:
Selecionamos o ano de 2007 para a visualização (pode ser alterado conforme preferência).

Criação do Scatter Plot:
Utilizamos seaborn para criar um scatter plot. Os parâmetros são:

x='gdpPercap': PIB per capita no eixo x.

y='lifeExp': Expectativa de vida no eixo y.

size='pop': Tamanho dos pontos representa a população.

hue='continent': Cor dos pontos representa o continente.

sizes=(20, 2000): Definição do tamanho mínimo e máximo dos pontos.

alpha=0.7: Transparência dos pontos.

palette='viridis': Paleta de cores para os continentes.

```python
# Importando bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados do Gapminder
url = 'https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv'
data = pd.read_csv(url)

# Visualizar as primeiras linhas do dataset
print(data.head())

# Selecionar o ano de 2007 para visualização
data_2007 = data[data['year'] == 2007]

# Criar scatter plot
plt.figure(figsize=(14, 8))
scatter = sns.scatterplot(
    data=data_2007,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    hue='continent',
    sizes=(20, 2000),
    alpha=0.7,
    palette='viridis'
)

# Adicionar títulos e labels
plt.title('Gapminder Data (2007)')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.legend(title='Continent', loc='upper left', bbox_to_anchor=(1, 1))
plt.xscale('log')  # Escala logarítmica para o PIB per capita
plt.grid(True)
plt.show()
```

# Observações Finais
Este scatter plot mostra quatro dimensões dos dados do Gapminder:

O PIB per capita no eixo x.
A expectativa de vida no eixo y.
O tamanho dos pontos representa a população.
A cor dos pontos representa o continente.
